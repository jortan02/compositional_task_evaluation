```python
import nltk
from nltk.corpus import wordnet as wn

# Download the WordNet data if not already done
nltk.download('wordnet')

# Get all the words in WordNet
all_words = list(wn.words())
```

```python
import statistics
statistics.mean([len(word) for word in all_words])
```
